```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dữ liệu
x_values = np.array([1.0279, 0.791, 0.6997, 1.4903, 1.8143, 0.7986, 1.1695, 1.227, 1.3592, 0.9603])
y_values = np.array([9.099999905, 9.700000048, 7.399999976, 10.79999995, 10, 8, 5.299999952, 5.300000072, 5.799999952, 5.600000024])

X = x_values.reshape(-1, 1)  
model = LinearRegression()
model.fit(X, y_values)

def predict_y(x):
    x_input = np.array(x).reshape(-1, 1)
    y_pred = model.predict(x_input)
    return y_pred[0]

plt.scatter(x_values, y_values, label='Actual Data')
plt.plot(x_values, model.predict(X), color='red', label='Floyd Complete')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

new_x_values = np.linspace(min(x_values), max(x_values), 10)
predicted_y_values = [predict_y(x) for x in new_x_values]

for i, x_val in enumerate(new_x_values):
    percentage_change = np.random.uniform(-0.2, 0.2)
    shifted_y = predicted_y_values[i] * (1 + percentage_change)

print(new_x_values)
print(shifted_y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dữ liệu
x_values = np.array([1.0279, 0.791, 0.6990, 1.4903, 1.8143, 0.7986, 1.1695, 1.227, 1.3592, 0.9603])
y_values = np.array([9.099999905, 9.700000048, 7.399999976, 10.79999995, 10, 8, 5.299999952, 5.300000072, 5.799999952, 5.600000024])

# Thực hiện linear regression
X = x_values.reshape(-1, 1)  # Chuyển đổi x_values thành ma trận (n, 1)
model = LinearRegression()
model.fit(X, y_values)

# Hàm dự đoán giá trị y khi nhập x
def predict_y(x):
    x_input = np.array(x).reshape(-1, 1)
    y_pred = model.predict(x_input)
    return y_pred[0]

# Vẽ biểu đồ dữ liệu thực tế và dự đoán
plt.scatter(x_values, y_values, label='Actual Data')
plt.plot(x_values, model.predict(X), color='red', label='Floyd Complete')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Tạo mảng 10 giá trị mới dao động trong khoảng min và max của x cũ
new_x_values = np.linspace(min(x_values), max(x_values), 12)
predicted_y_values = [predict_y(x) for x in new_x_values]

# In và hiển thị giá trị dự đoán cho các giá trị x mới
for i, x_val in enumerate(new_x_values):
    percentage_change = np.random.uniform(-0.2, 0.2)
    shifted_y = predicted_y_values[i] * (1 + percentage_change)
    print(f'{round(x_val, 4)}') 

print()

for i, x_val in enumerate(new_x_values):
    percentage_change = np.random.uniform(-0.2, 0.2)
    shifted_y = predicted_y_values[i] * (1 + percentage_change)
    print(f'{shifted_y}',)

plt.show()
```
